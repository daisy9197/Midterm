```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file
```

**First, I will use bokeh to plot**

```python
ad = pd.read_csv("/Users/apple/Documents/fall2018/python/Advertising.csv")
ad.info()
type(ad)
```

describe dataset
The type of this data is dataframe, it has 200 entries, 5 column.

```python
Sales = np.array(ad['Sales'])
TV = np.array(ad['TV'])
Radio = np.array(ad['Radio'])
Newspaper = np.array(ad['Newspaper'])

p1 = figure(title="TV,Radio,Newspaper VS Sales")
p1.grid.grid_line_alpha = 0.3
p1.yaxis.axis_label = "Sales"

p1.circle(TV,Sales,color='#A6CEE3',legend='TV')
p1.circle(Radio,Sales,color='#FB9A99',legend='Radio')
p1.circle(Newspaper,Sales,color='#33A02C', legend='News')
window_size = 30
window = np.ones(window_size)/float(window_size)

output_file("Sales.html",title="TV,Radio,Newspaper VS Sales")
show(p1)
```

**Second, I will use seaborn to plot**

```python
data = sns.load_dataset("tips")
data.info()
```

```python
#boxplot
sns.set_style ("whitegrid")
ax1 = sns.boxplot(x="day",y="total_bill",hue="smoker", data=data,palette="Set2")
ax1 = sns.swarmplot(x="day", y ="total_bill", data=data,color=".25")
```

```python
#violinplot
ax2 = sns.violinplot(x="day", y="total_bill", hue="smoker",data=data, palette="muted", split=True)
ax2 = sns.swarmplot(x="day",y="total_bill",data=data,color="white",edgecolor="grey")
```

```python
#kdeplot
sns.set(color_codes=True)
female = data.loc[data.sex == "Female"]
ax3 = sns.kdeplot(female.tip,female.total_bill,shade=True,shade_lowest=False)
```
